```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from sklearn.neighbors import NearestNeighbors
```

```python
data = pd.read_csv('data/encodedData.csv', header = 0, index_col = 0)
print(data)
```

```python
completeData = pd.read_csv('data/complete_dataset.csv', header = 0, index_col = None)
print(data)
```

```python
genres = sorted(list(set(completeData['genre'])))
print(genres)
```

```python
del genres[0]
```

```python
temp = set()
for x in genres:
    for y in x.split(","):
        temp.add((y.strip()))
genres = sorted(list(temp))
print(genres)
```

```python
genreMap = dict()
for i in range(len(completeData['Name'])):
    g = completeData['genre'][i]
    if type(g) is str:
        genreMap[completeData['Name'][i]] = map(lambda x : x.strip(), g.split(","))
    else:
        genreMap[completeData['Name'][i]] = g
```

```python
scoreMap = dict()
for i in range(len(completeData['Name'])):
    scoreMap[completeData['Name'][i]] = completeData['score'][i]
```

```python
user_genres = dict()

for x in data['Id']:
    user_genres[x] = [0 for y in range(len(genres))]

for i in range(len(completeData['Name'])):
    currentGenre = completeData['genre'][i]
    if type(currentGenre) is str:
        for genre in map(lambda x : x.strip(), currentGenre.split(",")):
            iVal = genres.index(genre)
            if iVal >= 0:
                user_genres[completeData['0'][i]][iVal] = 1
```

```python
print(user_genres[5250])
```

```python
genreEncode = dict()
for x in genres:
    genreEncode[x] = []

for key2 in sorted(user_genres.keys()):
    vector = user_genres[key2]
    i = 0
    for key in genres:
        genreEncode[key].append(vector[i])
        i += 1
```

```python
for x in sorted(genreEncode.keys()):
    data[x] = genreEncode[x]
```
